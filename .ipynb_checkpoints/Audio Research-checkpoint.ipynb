```python
from __future__ import print_function
from pyAudioAnalysis import audioBasicIO
from pyAudioAnalysis import audioFeatureExtraction
import matplotlib.pyplot as plt
import numpy as np
```

```python
frame_size = 0.0100
frame_step = 0.050
```

```python
import os
import random
import sys
reload(sys)
sys.setdefaultencoding("utf-8")

def getRandomFile(path):
  """
  Returns a random filename, chosen among the files of the given path.
  """
  files = os.listdir(path)
  index = random.randrange(0, len(files))
  return path+files[index]
```

```python
#load pill data

#1 pill
[Fs1, x1] = audioBasicIO.readAudioFile(getRandomFile("audio/pills_1/"));
F1 = audioFeatureExtraction.stFeatureExtraction(x1, Fs1, frame_size*Fs1, frame_step*Fs1);

#10 pills
[Fs2, x2] = audioBasicIO.readAudioFile(getRandomFile("audio/pills_10/"));
F2 = audioFeatureExtraction.stFeatureExtraction(x1, Fs1, frame_size*Fs1, frame_step*Fs1);

#25 pills
[Fs3, x3] = audioBasicIO.readAudioFile(getRandomFile("audio/pills_25/"));
F3 = audioFeatureExtraction.stFeatureExtraction(x3, Fs3, frame_size*Fs3, frame_step*Fs3);

#50 pills
[Fs4, x4] = audioBasicIO.readAudioFile(getRandomFile("audio/pills_50/"));
F4 = audioFeatureExtraction.stFeatureExtraction(x4, Fs4, frame_size*Fs4, frame_step*Fs4);
```

```python
def plot(feature, title):
    fig, ax = plt.subplots()
    ax.plot(F1[feature,:], label='1 pill')
    ax.plot(F2[feature,:], label='10 pills')
    ax.plot(F3[feature,:], label='25 pills')
    ax.plot(F4[feature,:], label='50 pills')

    # Now add the legend with some customizations.
    legend = ax.legend(loc='upper center', shadow=True)

    # The frame is matplotlib.patches.Rectangle instance surrounding the legend.
    frame = legend.get_frame()
    frame.set_facecolor('0.90')

    plt.xlabel('Frame no'); plt.ylabel(title); 

    plt.title(title)
    plt.show()
```

```python
plot(0, "Zero Crossing Rate") #The rate of sign-changes of the signal during the duration of a particular frame.
```

```python
plot(1, "Energy") #The sum of squares of the signal values, normalized by the respective frame length.
```

```python
plot(2, "Entropy of Energy") #The entropy of sub-frames' normalized energies. It can be interpreted as a measure of abrupt changes.
```

```python
plot(3, "Spectral Centroid") #The center of gravity of the spectrum.
```

```python
plot(4, "Spectral Spread") #The second central moment of the spectrum.
```

```python
plot(5, "Spectral Entropy") #Entropy of the normalized spectral energies for a set of sub-frames.
```

```python
plot(6, "Spectral Flux") #the squared difference between the normalized magnitudes of the spectra of the two successive frames.
```
